# **Student Result Management System**

The Student Result Management System is a Python-based project that helps schools, colleges, and institutes manage student performance easily. It allows teachers to enter each student’s marks in multiple subjects, automatically calculates the total and average using NumPy, and assigns a grade based on performance. The program instantly shows the student’s individual report and later displays a summary of all students’ results together. It uses concepts like variables, lists, loops, control flow, and mathematical functions to process data efficiently. This project saves time, reduces manual calculation errors, and provides a quick, accurate way to manage academic results.

| Category              | Example                           | Use                                          |
| --------------------- | --------------------------------- | -------------------------------------------- |
| **Variables**         | `name`, `math`, `total`, `grade`  | Data store karne ke liye                     |
| **Data Types**        | `string`, `float`, `list`, `dict` | Different kinds of data handle karne ke liye |
| **Loop**              | `while`, `for`                    | Repeat tasks like taking multiple students   |
| **Control Flow**      | `if`, `elif`, `else`, `break`     | Decision making aur loop control             |
| **NumPy Library**     | `np.sum()`, `np.mean()`           | Fast mathematical operations                 |
| **Functions**         | `input()`, `print()`, `append()`  | User input, output aur data storage          |
| **String Formatting** | `f"{variable}"`                   | Beautiful output display                     |


```python
import numpy as np

print(" Welcome to Student Result Management System ")

students = []  # List to store all student data

while True:
    name = input("\nEnter Student Name (or type 'done' to finish): ")
    if name.lower() == "done":
        break

    # Get marks
    math = float(input("Enter Math Marks: "))
    science = float(input("Enter Science Marks: "))
    english = float(input("Enter English Marks: "))
    computer = float(input("Enter Computer Marks: "))

    marks = np.array([math, science, english, computer])

    # Calculate total and average
    total = np.sum(marks)
    average = np.mean(marks)

    # Grade logic
    if average >= 80:
        grade = "A+"
    elif average >= 70:
        grade = "A"
    elif average >= 60:
        grade = "B"
    elif average >= 50:
        grade = "C"
    else:
        grade = "Fail"

    # Show result immediately
    print("\n Student Report:")
    print(f"Name: {name}")
    print(f"Marks: {marks}")
    print(f"Total Marks: {total}")
    print(f"Average Marks: {average:.2f}")
    print(f"Grade: {grade}")

    # Save student record
    students.append({
        "Name": name,
        "Marks": marks,
        "Total": total,
        "Average": average,
        "Grade": grade
    })
print("\n All results generated successfully!")
```
